```python
names = ['Aryan', 'dhanu', 'sammruddhi', 'swaraj', 'Sukanya', 'shivraj']
```

```python
names
```

```python
# create a new list with names which starts with 's' or 'S'
```

```python
for i in names:
    if i.lower().startswith('s'):
        print(i)
```

```python
new = [i for i in names if i.lower().startswith('s')]
new
```

```python
for i in names:
    if i[0] == 's' or i[0] == 'S':
        print(i)
```
